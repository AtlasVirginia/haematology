```python
import pandas as pd
df = pd.read_csv('standard.csv')
df2 = pd.read_csv('provided.csv')
```

```python
merged = pd.merge(left=df, right=df2, how='left' , left_on='test_name', right_on='test_name')
sex = 'f'
if sex == 'f':
    print(' patient sex is female')
    merged['bigger_than_f'] = df['high_female'] < df2['provided']
    merged['smaller_than_f'] = df['low_female'] > df2['provided']
    merged['equal_f'] = merged['smaller_than_f'] == merged['bigger_than_f']
elif sex == 'm':
    print('patient sex is male')
    merged['bigger_than_m'] = df['high_male'] < df2['provided']
    merged['smaller_than_m'] = df['low_male'] > df2['provided']
    merged['equal_m'] = merged['smaller_than_m'] == merged['bigger_than_m']


merged.to_csv('blood.csv')
merged
```
